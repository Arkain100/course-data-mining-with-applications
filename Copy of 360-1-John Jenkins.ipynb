Write some code that allows the user to play Rock, Paper, Scissors against the computer.

The game should play at least five rounds and keep playing until the user has a 40% win rate.

Within each round, the program should
* ask the user which they pick Rock, Paper, or Scissors
* use a random integer generator to determine which value the computer picks between Rock, Paper, Scissors
* state both values and if the user won (if you tie, you both win)
* record the total number of rounds you have played
* record the total number of wins you have

At the end of the game, report back the total number of rounds and your win percentage. *(7 points)*

```python
i=1
win=0
lose=0
t=0
while (i<6 or t<.4):
  nun=random.randint(1,3)
  user=input("Rock, paper, or scissors? ")
  if user.lower()=='rock':
    play=1
  elif user.lower()=='paper':
    play=2
  else:
    play=3
  if play-nun==1:
    win=win+1
  elif nun-play==1:
    lose=lose+1
  elif play-nun==2:
    lose=lose+1
  elif nun-play==2:
    win=win+1
  elif nun==play:
    win=win+1
    lose=lose+1
  if nun==1:
    state="rock"
  elif nun==2:
    state="paper"
  else:
    state="scissors"
  print("User: ",user, "Game: ",state)
  if win==0:
    t=0
  else:
    t=win/i
  i=i+1
print("Rounds: ",i-1)
print("Wins: ",win)
print("Win percentage: ",t*100,"%")
```
